```python
def output_values(tr_results):
    l = []
    i = 0
    while i < len(tr_results):
        if tr_results[i] == 'passed':
            l.append(1)
        else:
            l.append(0)
        i += 1
    
    return l
```

```python
def batchstop4(commits):
    global batch_total
    
    l = len(commits)//2
    
    if 0 in commits:
        
        if len(commits) <= 4:
            return len(commits)
        
        return 1 + batchstop4(commits[:l]) + batchstop4(commits[l:])
    
    return 1
    
```

```python
def batchbisect(commits):
    global batch_total
    
    l = len(commits)//2
    
    if len(commits) == 1:
        return 1
    
    if 0 in commits:
        return 1 + batchbisect(commits[:l]) + batchbisect(commits[l:])
    
    return 1
```

```python
def runbatch(batch):

    if 0 in batch:
        return False
    return True
```

```python
def linear_update(res, batch, batch_size, factor, min_size=1):
    fails_per = 100*batch.count(0)/len(batch)
    
    if res == 0:

        if (fails_per < 20) & (batch[-1] == 1):
            return batch_size

        #if batch has more than 20% failures then definitely split
        if batch_size <= factor:
            return min_size
        else:
            batch_size -= factor


    else:
        
        if batch_size >= 16:
            return batch_size
        else:
            batch_size += factor
    
    return batch_size
```

```python
def exp_update(res, batch, batch_size, factor, min_size=1):
    fails_per = 100*batch.count(0)/len(batch)
    if res == 0:

        if (fails_per < 20) & (batch[-1] == 1):
            return batch_size
        
        #if batch has more than 20% failures then definitely split
        if batch_size <= factor:
            return min_size
        else:
            batch_size //= factor


    else:
        
        if batch_size >= 16:
            return batch_size
        else:
            batch_size *= factor
    
    return batch_size
```

```python
def stagger_update(res, batch, batch_size, factor, min_size=1):
    
    fails_per = 100*batch.count(0)/len(batch)
    
    if res == 0:
        
        if batch_size <= factor:
            return min_size
        
        if (fails_per < 10) & (batch[-1] == 1):
            return batch_size
        if (fails_per < 20) & (batch[-1] == 1):
            return batch_size - factor
        if (fails_per < 50) & (batch[-1] == 1):
            return batch_size // factor
        else:
            batch_size = 1


    else:
        
        if batch_size >= 16:
            return 16
        else:
            batch_size = min(batch_size * factor, 16)
    
    return batch_size
```

```python
def random_linear(res, batch, batch_size, factor, min_size=1):
    fails_per = 100*batch.count(0)/len(batch)
    
    
    
    if res == 0:
                
        factor = random.randint(1, batch_size)

        if (fails_per < 20) & (batch[-1] == 1):
            return batch_size
        
        #if batch has more than 20% failures then definitely split
        if batch_size <= factor:
            return min_size
        else:
            batch_size -= factor


    else:
        
        factor = random.randint(2, 4)
        
        if batch_size >= 16:
            return batch_size
        else:
            batch_size += factor
    
    return batch_size
```

```python
def random_exp(res, batch, batch_size, factor, min_size=1):
    fails_per = 100*batch.count(0)/len(batch)
    
    if res == 0:        
        factor = random.randint(1, batch_size)
        
        if (fails_per < 20) & (batch[-1] == 1):
            return batch_size
        
        #if batch has more than 20% failures then definitely split
        if batch_size <= factor:
            return min_size
        else:
            batch_size //= factor


    else:
        
        factor = random.randint(2, 4)
        
        if batch_size >= 16:
            return batch_size
        else:
            batch_size *= factor
    
    return batch_size
```

```python
def batchstop4_dynamic(commits):
    global batch_total
    
    batch_size = 16
    length = len(commits)
    builds = 0
    i = 0
    delays = []
    
    while i < length:
        batch = commits[i:i+batch_size]
        fails_per = 100*batch.count(0)/batch_size
        delays.extend([batch_size-x-1 for x in range(batch_size)])
        
        batch_total = 0
        res = batchstop4(batch)
        builds += batch_total
        print('Building {} {} {} {}'.format(batch_size, batch, batch_total, builds))
        
        i += batch_size
        
        #updating batch size
        batch_size = linear_update(res, batch)
    
    return builds, delays
```
