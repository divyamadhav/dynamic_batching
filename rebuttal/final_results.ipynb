```python
import pandas as pd
```

```python
baseline_dynamic = pd.read_csv('baseline_dynamic_batching_timed.csv')
ci_skip_baseline_dynamic = pd.read_csv('ci_skipped_baseline_dynamic_batching_timed.csv')
baseline_static = pd.read_csv('static_batching_duration.csv')
ci_skip_baseline_static = pd.read_csv('ci_skipped_static_batching_duration.csv')

project_list = list(set(baseline_static['project'].tolist()))
```

```python
print(len(baseline_dynamic))
baseline_dynamic = baseline_dynamic[(baseline_dynamic['project'].isin(project_list))]
print(len(baseline_dynamic))

baseline_dynamic = baseline_dynamic.rename(columns={'duration': 'total_duration'})
baseline_dynamic.columns
```

```python
our_dynamic = pd.read_csv('lwd_time_saved.csv')
our_ci_dynamic = pd.read_csv('ci_skipped_lwd_time_saved.csv')
mfu = pd.read_csv('mfu_time_saved.csv')
```

```python
project_list = list(set(our_dynamic['project'].tolist()))
```

```python
our_dynamic['ci_skip'] = [0]*len(our_dynamic)
our_ci_dynamic['ci_skip'] = [1]*len(our_ci_dynamic)
baseline_static['ci_skip'] = [0]*len(baseline_static)
ci_skip_baseline_dynamic['ci_skip'] = [1]*len(ci_skip_baseline_dynamic)
ci_skip_baseline_static['ci_skip'] = [1]*len(ci_skip_baseline_static)
```

```python
baseline_dynamic['method'] = ['baseline_dynamic']*len(baseline_dynamic)
our_dynamic['method'] = ['new_dynamic']*len(our_dynamic)
our_ci_dynamic['method'] = ['new_dynamic']*len(our_ci_dynamic)
baseline_static['method'] = ['baseline_static']*len(baseline_static)
ci_skip_baseline_dynamic['method'] = ['baseline_dynamic']*len(ci_skip_baseline_dynamic)
ci_skip_baseline_static['method'] = ['baseline_static']*len(ci_skip_baseline_static)
```

```python
final_results = pd.DataFrame()
final_results = final_results.append(baseline_dynamic)
final_results = final_results.append(ci_skip_baseline_dynamic)
final_results = final_results.append(our_dynamic)
final_results = final_results.append(our_ci_dynamic)
final_results = final_results.append(mfu)
final_results = final_results.append(baseline_static)
final_results = final_results.append(ci_skip_baseline_static)
```

```python
final_results = final_results.loc[:, ~final_results.columns.str.contains('^Unnamed')]
```
