```python
def output_values(tr_results):
    l = []
    i = 0
    while i < len(tr_results):
        if tr_results[i] == 'passed':
            l.append(1)
        else:
            l.append(0)
        i += 1
    
    return l
```

```python
def batchstop4(commits, duration):
    global batch_duration
    
    l = len(commits)//2
    
    if 0 in commits:
        
        if len(commits) < 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1+len(commits)
        
        if len(commits) == 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1 + len(commits)
        
        batch_duration += duration[:l][-1]
        batch_duration += duration[l:][-1]
                
        return 1 + batchstop4(commits[:l], duration[:l]) + batchstop4(commits[l:], duration[l:])
    
    batch_duration += duration[-1]
    return 1
```

```python
def batchbisect(commits, duration):
    global batch_duration
        
    l = len(commits)//2
    
    batch_duration += duration[-1]
    
    if len(commits) == 1:
        return 1
    
    if 0 in commits:
        return 1 + batchbisect(commits[:l], duration[:l]) + batchbisect(commits[l:], duration[:l])
    
    return 1
```

```python
algorithms = {'BATCHBISECT':batchbisect, 'BATCHSTOP4':batchstop4, 'BATCHDIVIDE4':batchdivide4}
```

```python
def mfu(results, factor, algorithm, duration, min_size = 1, max_size = 16):
    
    global batch_duration
    
    max_batch_size = max_size
    min_batch_size = min_size
    batch_sizes = []
    total_duration = 0
    
    
    i = 0
    builds = 0
    length = len(results)
    
    cur_batch_size = min_batch_size
    
    #print(duration)
    
    while i < length:
        
        batch_duration = 0
        
        batch = results[i:i+cur_batch_size]
        cur_duration = duration[i:i+cur_batch_size]
        
        i = i+cur_batch_size
        batch_sizes.append(cur_batch_size)
        
        batch_total = algorithm(batch, cur_duration)
        builds += batch_total
        total_duration += batch_duration
                
        if 0 in batch:
            res = 0
        else:
            res = 1
        
        fails_per = 100*batch.count(0)/len(batch)
        #print(cur_batch_size, batch, batch_total, fails_per)
        
        if res == 0:
            
            if fails_per >= 50:
                cur_batch_size = min_batch_size
        
            if cur_batch_size <= factor:
                cur_batch_size = min_batch_size
                
            elif (fails_per < 10) & (batch[-1] == 1):
                pass
        
            elif fails_per < 20:
                cur_batch_size -= factor

            elif fails_per < 50:
                cur_batch_size //= factor

            else:
                cur_batch_size = Counter(batch_sizes).most_common(1)[0][0]
                print('Most common {}'.format(cur_batch_size))
        
        else:
            
            if cur_batch_size >= 16:
                cur_batch_size = max_batch_size
            else:
                cur_batch_size = min(cur_batch_size * factor, 16)
        
    
    builds_saved = 100-(100*builds/length)
    return builds, builds_saved, total_duration
```
