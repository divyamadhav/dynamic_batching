```python
def output_values(tr_results):
    l = []
    i = 0
    while i < len(tr_results):
        if tr_results[i] == 'passed':
            l.append(1)
        else:
            l.append(0)
        i += 1
    
    return l
```

```python
def batchstop4(commits, duration):
    global batch_duration
    
    
    l = len(commits)//2
    
    if 0 in commits:
        
        if len(commits) < 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            
            return 1+len(commits)
        
        
        if len(commits) == 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1 + len(commits)
        
        batch_duration += duration[:l][-1]
        batch_duration += duration[l:][-1]
        
        
        return 1 + batchstop4(commits[:l], duration[:l]) + batchstop4(commits[l:], duration[l:])
    
    batch_duration += duration[-1]
    return 1
```

```python
def batchbisect(commits, duration):
    global batch_duration
    
    
    l = len(commits)//2
    
    batch_duration += duration[-1]
    
    if len(commits) == 1:
        return 1
    
    if 0 in commits:
        return 1 + batchbisect(commits[:l], duration[:l]) + batchbisect(commits[l:], duration[:l])
    
    return 1
```

```python
def dynamic_batch(commits, duration, testall_duration, print_flag=0):
    
    global linear_update, exp_update, random_linear, random_exp, freq_update, random_random, half_exp_update
    global batchbisect, batchstop4, batchdivide4
    global batch_total
    global batch_duration
    
    
    rows = []
    batch_size = 16
    length = len(commits)
    builds = 0
    i = 0
    delays = []
    lines = 0
    min_allowed = 1
    max_allowed = 16
    
    algorithms = {'BATCHDIVIDE4':batchdivide4, 'BATCHBISECT': batchbisect, 'BATCHSTOP4':batchstop4}
    methods = {'linear': linear_update, 'exponential':exp_update, 'stagger':stagger_update, 
               'random_linear':random_linear, 'random_exponential':random_exp, 'random_random':random_random,
               }

    for m in methods:
        for alg in algorithms:
            
            
            batch_size = 16
            
            if m == 'linear':
                factors = [1, 2, 3, 4]
                
                for f in factors:
                    
                    
                    builds = 0
                    delays = []
                    i = 0
                    batch_size = 16
                    
                    total_duration = 0

                    while i < length:
                                                
                        batch = commits[i:i+batch_size]
                        build_time = duration[i:i+batch_size]
                        
                        if len(batch) != len(build_time):
                            print('This is ridiculous for {} {} {}'.format(m, f, alg))
                            break

                        delays.extend([batch_size-x-1 for x in range(batch_size)])
                        
                        batch_duration = 0
                        
                        batch_total = algorithms[alg](batch, build_time)
                        total_duration += batch_duration
                        
                        builds += batch_total
                        
#                         print(alg, m)
#                         print(batch_size, batch, batch_total, builds)

                        i += batch_size
                        
                        if 0 in batch:
                            res = 0
                        else:
                            res = 1
                        #updating batch size
                        if alg == 'BATCHBISECT':
                            batch_size = methods[m](res, batch, batch_size, f)
                        else:
                            batch_size = methods[m](res, batch, batch_size, f, 4)
                        

                    saved = 100-(100*builds/length)
                    fr = 100*results.count(0)/length
                    time_saved = 100-(100*total_duration/testall_duration)
                    rows.append([p_name, str(alg), '-1', str(m), f, total_duration, testall_duration, time_saved, builds, saved, length])
                    lines += 1
                    
            
            elif (m == 'exponential') or (m == 'stagger') or (m == 'half_exp'):
                factors = [2, 3]
                
                for f in factors:
            
                    builds = 0
                    delays = []
                    i = 0
                    batch_size = 16
                    
                    total_duration = 0

                    while i < length:
                        batch = commits[i:i+batch_size]
                        build_time = duration[i:i+batch_size]
                        
                        batch_duration = 0

                        delays.extend([batch_size-x-1 for x in range(batch_size)])

                        batch_total = algorithms[alg](batch, build_time)
                        builds += batch_total
                        
                        total_duration += batch_duration
#                         print(alg, m)
#                         print(batch_size, batch, batch_total, builds)

                        i += batch_size
                        if 0 in batch:
                            res = 0
                        else:
                            res = 1

                        #updating batch size
                        if alg == 'BATCHBISECT':
                            batch_size = methods[m](res, batch, batch_size, f)
                        else:
                            batch_size = methods[m](res, batch, batch_size, f, 4)

                    saved = 100-(100*builds/length)
                    fr = 100*results.count(0)/length
                    time_saved = 100-(100*total_duration/testall_duration)
                    rows.append([p_name, str(alg), '-1', str(m), f, total_duration, testall_duration, time_saved, builds, saved, length])
                    lines += 1
            
            else:
                
                builds = 0
                delays = []
                i = 0
                total_duration = 0
                batch_size = 16

                while i < length:
                    batch = commits[i:i+batch_size]
                    build_time = duration[i:i+batch_size]
                    
                    batch_duration = 0

                    delays.extend([batch_size-x-1 for x in range(batch_size)])

                    batch_total = algorithms[alg](batch, build_time)
                    builds += batch_total
                    
                    total_duration += batch_duration
#                     print(alg, m)
#                     print(batch_size, batch, batch_total, builds)

                    i += batch_size
                    if 0 in batch:
                        res = 0
                    else:
                        res = 1

                    #updating batch size
                    if alg == 'BATCHBISECT':
                        batch_size = methods[m](res, batch, batch_size, f)
                    else:
                        batch_size = methods[m](res, batch, batch_size, f, 4)

                saved = 100-(100*builds/length)
                fr = 100*results.count(0)/length
                time_saved = 100-(100*total_duration/testall_duration)
                rows.append([p_name, str(alg), '-1', str(m), -1, total_duration, testall_duration, time_saved, builds, saved, length])
                lines += 1

            #print([p_name, alg, '-1', m, saved, length])
    #print(lines)
    return rows
            
```

```python
i = 0

standard_rows = []
ci_skip_rows = []


for p in project_list:
    
    print(p)
    p_name = p
    
    if (p_name == 'BuildCraft-BuildCraft.csv') | (p_name == 'RS485-LogisticsPipes.csv') | (p_name == 'datastax-java-driver.csv') :
        continue
        
    data = pd.read_csv('../dynamic batching/ci_skip_data/extracted_project_travis/' + p_name)
    
    branch_type = data['git_branch'].tolist()
    if 'master' in branch_type:
        data = data[ data['git_branch'] == 'master']
    else:
        data = data[ data['git_branch'] == 'trunk']
        
    total_length = len(data) - 100
    run_data = data.iloc[100:]
    if len(run_data) == 0:
        print(i)
        i += 1
        continue
    
    time_reqd = run_data['tr_duration'].tolist()
    results = output_values(run_data['tr_status'].tolist())
    
    testall_duration = sum(run_data['tr_duration'].tolist())
    
    rows = dynamic_batch(results, time_reqd, testall_duration)
    standard_rows.extend(rows)
    
    
    
    run_data = data[ data.verdict == 0]
    run_data = run_data.iloc[100:]
    
    
    if len(run_data) == 0:
        print(i)
        i += 1
        continue
        
    results = output_values(run_data['tr_status'].tolist())
    time_reqd = run_data['tr_duration'].tolist()
    rows = dynamic_batch(results, time_reqd, testall_duration, print_flag=1)
    ci_skip_rows.extend(rows)
    
    
    
```
