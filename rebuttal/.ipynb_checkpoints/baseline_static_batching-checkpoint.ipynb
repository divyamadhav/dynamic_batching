```python
def batchbisect(commits, duration):
    global batch_duration
    
    l = len(commits)//2
    
    batch_duration += duration[-1]
    
    if len(commits) == 1:
        return 1
    
    if 0 in commits:
        return 1 + batchbisect(commits[:l], duration[:l]) + batchbisect(commits[l:], duration[l:])
    
    return 1
```

```python
def batchstop4(commits, duration):
    global batch_duration
    
    l = len(commits)//2
    
    if 0 in commits:
        
        if len(commits) < 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1+len(commits)
        
        if len(commits) == 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1 + len(commits)
        
        batch_duration += duration[:l][-1]
        batch_duration += duration[l:][-1]
                
        return 1 + batchstop4(commits[:l], duration[:l]) + batchstop4(commits[l:], duration[l:])
    
    batch_duration += duration[-1]
    return 1
```

```python
def output_values(tr_results):
    l = []
    i = 0
    while i < len(tr_results):
        if tr_results[i] == 'passed':
            l.append(1)
        else:
            l.append(0)
        i += 1
    
    return l
```

```python
def batching(project, data, algorithm, string):
    global batch_total
    global batch_duration
    
    p_name = project
    
    testall_duration = sum(data['tr_duration'].tolist())
    
    rows = []
    
    results = output_values(data['tr_status'].tolist())
    duration = data['tr_duration'].tolist()
    
    length = len(results)
    
    if string == 'BATCHBISECT':
        batch_sizes = [2, 4, 8, 16]
    elif string == 'BATCHSTOP4':
        batch_sizes = [4, 8, 16]
    else:
        batch_sizes = [4, 8, 16]
    
    
       
    for b in batch_sizes:
        print('Building {} {} {}'.format(algorithm, b, testall_duration))
        builds = 0
        delays = []
        
        total_duration = 0
        i = 0
        while i < length:
            
            if length - i > b:
                batch = results[i:i+b]
                cur_duration = duration[i:i+b]
                delays.extend([b-x-1 for x in range(b)])
            else:
                batch = results[i:]
                cur_duration = duration[i:]
                delays.extend([b-x-1 for x in range(length-i)])
            batch_duration = 0
            batch_total = algorithm(batch, cur_duration)
            builds += batch_total
            total_duration += batch_duration
            
            #print('Executed \t {} \t {} \t {} \t {}'.format(builds, total_duration, batch,  cur_duration))
            #print(b, cur_duration, batch_duration, total_duration)
            
            i += b
        
        builds_saved = 100-(100*builds/length)
        median_delay = median(delays)
        
        per_saved_time = 100-(100*total_duration/testall_duration)
        rows.append([p_name, string, b, total_duration, testall_duration, per_saved_time, builds_saved, builds, length])
    
    return rows
```

```python
discard_list = []
ci_skipped_result_rows = []
result_rows = []

# project_list = ['./apache-pdfbox.csv']


for p in project_list:
    p_name = p
    try:
        data = pd.read_csv('../dynamic batching/ci_skip_data/extracted_project_travis/'+p_name)
    except:
        print(p)
        continue
    
    branch_type = data['git_branch'].tolist()
    if 'master' in branch_type:
        data = data[ data['git_branch'] == 'master']
    else:
        data = data[ data['git_branch'] == 'trunk']
        
    
    norm_data = data.iloc[100:]
    
    if len(norm_data) == 0:
        print('Could not continue for {}'.format(p))
        discard_list.append(p)
        continue
        
    rows = batching(p, norm_data, batchbisect, 'BATCHBISECT')
    result_rows.extend(rows)
    
    rows = batching(p, norm_data, batchstop4, 'BATCHSTOP4')
    result_rows.extend(rows)
    
    
    rows = batching(p, norm_data, batchdivide4, 'BATCHDIVIDE4')
    result_rows.extend(rows)
    
    
    data = data[ data['verdict'] == 0]
    data = data.iloc[100:]
    
    if len(data) == 0:
        print('Could not continue for {}'.format(p))
        discard_list.append(p)
        continue
        
    rows = batching(p, data, batchbisect, 'BATCHBISECT')
    ci_skipped_result_rows.extend(rows)
    
    rows = batching(p, data, batchstop4, 'BATCHSTOP4')
    ci_skipped_result_rows.extend(rows)
    
    rows = batching(p, data, batchdivide4, 'BATCHDIVIDE4')
    ci_skipped_result_rows.extend(rows)

    
```
