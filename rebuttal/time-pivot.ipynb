```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
bs = pd.read_csv('static_batching_duration.csv')
bd = pd.read_csv('baseline_dynamic_batching_timed.csv')
```

```python
project_list = list(set(bs['project'].tolist()))
```

```python
bs_off = bs[bs['testall_duration'].isna()]['project'].tolist()
bd_off = bd[(bd['testall_duration'] == 0)]['project'].tolist()
new_project_list = list(set(project_list).difference(bs_off))
new_project_list = list(set(new_project_list).difference(bd_off))
```

```python
ld = pd.read_csv('lwd_time_saved.csv')
mfu = pd.read_csv('mfu_time_saved.csv')

ld = ld.replace('random_random', 'random')
ld = ld.replace('stagger', 'mixed')
mfu = mfu.replace('stagger_mfu', 'MFU')
```

```python
bd = bd[ bd['project'].isin(new_project_list)]
bs = bs[ bs['project'].isin(new_project_list)]
ld = ld[ ld['project'].isin(new_project_list)]
mfu = mfu[ mfu['project'].isin(new_project_list)]
```

```python
bd['time_saved'] = bd['time_saved'].apply(lambda x: 100-x)
ld['time_saved'] = ld['time_saved'].apply(lambda x: 100-x)
bs['time_saved'] = bs['time_saved'].apply(lambda x: 100-x)
mfu['time_saved'] = mfu['time_saved'].apply(lambda x: 100-x)
```

```python
ld['final_methods'] = ld['algorithm'] + ' ' + ld['method'] + ' ' + ld['factor'].astype(str)
mfu['final_methods'] = mfu['algorithm'] + ' ' + mfu['update_method'] + ' ' + mfu['factor'].astype(str)
bd['final_methods'] = 'Baseline Dynamic' + ' ' + bd['algorithm']
bs['final_methods'] = bs['algorithm'] + ' ' + bs['batch_size'].astype(str)
```

```python
combined = pd.concat([ld, mfu, bd, bs])
```

```python
final_methods = list(set(combined['final_methods'].tolist()))
```

```python
flag = True
```

```python
for f in final_methods:
    
    combined_subsection = combined[ combined['final_methods'] == f]
    combined_subsection = combined_subsection.sort_values(by='project')
    
    if flag:
        df['project'] = combined_subsection['project']
        flag = False
    
    df[f] = combined_subsection['time_saved'].tolist()
```
