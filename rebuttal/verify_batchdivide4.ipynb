```python
def batchstop4(commits, duration):
    global batch_duration
    
    l = len(commits)//2
    
    if 0 in commits:
        
        if len(commits) < 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1 + len(commits)
        
        if len(commits) == 4:
            for i in range(len(duration)):
                batch_duration += duration[i]
            return 1 + len(commits)
        
        batch_duration += duration[:l][-1]
        batch_duration += duration[l:][-1]
        
        print('Bisected into \t {} {}'.format(commits[:l], commits[l:]))
        
        return 1 + batchstop4(commits[:l], duration[:l]) + batchstop4(commits[l:], duration[l:])
    
    batch_duration += duration[-1]
    return 1
```
