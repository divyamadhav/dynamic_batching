```python
import pandas as pd
```

```python
data = pd.read_csv('ci_skip_all_results.csv')
```

```python
lwd = data[ data['method'] == 'new_dynamic']
bd = data[data['method'] == 'baseline_dynamic']
```

```python
lwd['final_methods'] = lwd['update_method'] + '-' + lwd['factor'].astype(str)
```

```python
final_methods = list(set(lwd['final_methods'].tolist()))
```

```python
algorithms = ['BATCHBISECT', 'BATCHSTOP4', 'BATCHDIVIDE4']
```

```python
project_list = set({})
```

```python
for alg in algorithms:
    
    bd_alg = bd[ (bd['algorithm'] == alg) & (bd['ci_skip'] == 0)]
    lwd_alg = lwd[ (lwd['algorithm'] == alg) & (lwd['ci_skip'] == 0)]
    
    bd_alg = bd_alg.sort_values(by='project')
    
    for f in final_methods:
        
        m_lwd = lwd_alg[ lwd_alg['final_methods'] == f]
        m_lwd = m_lwd.sort_values(by='project')
        
        for i in range(50):
            row_m = m_lwd.iloc[i]
            row_bd = bd_alg.iloc[i]
            
            if row_m['builds_saved'] > row_bd['builds_saved']:
                if row_m['time_saved'] < row_bd['time_saved']:
                    print(row_m['project'], row_bd['project'])
                    project_list.add(row_m['project'])
        
        print('{} done\n'.format(f))
    print('{} done\n'.format(alg))
```
