```python
import pandas as pd
```

```python
baseline_dynamic = pd.read_csv('baseline_dynamic_batching.csv')
ci_skipped_baseline_dynamic = pd.read_csv('ci_skipped_baseline_dynamic_batching.csv')
our_dynamic = pd.read_csv('new_dynamic_batching.csv')
our_ci_dynamic = pd.read_csv('ci_skipped_dynamic_batching.csv')
baseline_static = pd.read_csv('static_baseline_batching.csv')
baseline_ci_static = pd.read_csv('ci_skipped_static_baseline_batching.csv')
```

```python
our_dynamic['ci_skip'] = [0]*len(our_dynamic)
our_ci_dynamic['ci_skip'] = [1]*len(our_ci_dynamic)
baseline_static['ci_skip'] = [0]*len(baseline_static)
baseline_ci_static['ci_skip'] = [1]*len(baseline_ci_static)
```

```python
baseline_dynamic['method'] = ['baseline_dynamic']*len(baseline_dynamic)
ci_skipped_baseline_dynamic['method'] = ['baseline_dynamic']*len(ci_skipped_baseline_dynamic)
our_dynamic['method'] = ['new_dynamic']*len(our_dynamic)
our_ci_dynamic['method'] = ['new_dynamic']*len(our_ci_dynamic)
baseline_static['method'] = ['baseline_static']*len(baseline_static)
baseline_ci_static['method'] = ['baseline_static']*len(baseline_ci_static)
```

```python
final_results = pd.DataFrame()
final_results = final_results.append(baseline_dynamic)
final_results = final_results.append(ci_skipped_baseline_dynamic)
final_results = final_results.append(our_dynamic)
final_results = final_results.append(our_ci_dynamic)
final_results = final_results.append(baseline_static)
final_results = final_results.append(baseline_ci_static)
```
